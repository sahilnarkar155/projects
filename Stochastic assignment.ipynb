````python
import numpy as np
import matplotlib.pyplot as plt

def poisson_process(lmbda, T):
    # generate time intervals
    t = np.cumsum(np.random.exponential(1/lmbda, int(lmbda*T)))
    # return times of events
    return t[t <= T]

# example usage
lmbda = 2  # average rate of 2 events per unit time
T = 10  # time horizon of 10 units
events = poisson_process(lmbda, T)
print(events)
plt.eventplot(events)
plt.show()
````

````python
def superposition_process(lmbda1, lmbda2, T):
    events1 = poisson_process(lmbda1, T)
    events2 = poisson_process(lmbda2, T)
    return np.sort(np.concatenate((events1, events2)))

# example usage
lmbda1 = 2  # average rate of 2 events per unit time for process 1
lmbda2 = 3  # average rate of 3 events per unit time for process 2
T = 10  # time horizon of 10 units
events = superposition_process(lmbda1, lmbda2, T)
print(events)
plt.eventplot(events)
plt.show()
````

````python
def compound_process(lmbda1, lmbda2, T, p):
    events1 = poisson_process(lmbda1, T)
    events2 = poisson_process(lmbda2, T)
    events2_thinned = thinning_process(events2, p)
    return np.sort(np.concatenate((events1, events2_thinned)))

# example usage
lmbda1 = 1  # average rate of 1 customer per minute for regular customers
lmbda2 = 2  # average rate of 2 customers per minute for VIP customers
T = 60  # time horizon of 60 minutes
p = 0.5  # probability of a VIP customer becoming a regular customer
events = compound_process(lmbda1, lmbda2, T, p)
print(events)
plt.eventplot(events)
plt.show()
````
